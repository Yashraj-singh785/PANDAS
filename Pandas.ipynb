 ## What is Pandas?

To master pandas, you must first understand the two workhorses that power it: the Series and the DataFrame. While NumPy (which we covered in Chapter 4) is great for raw numerical crunching, it lacks a native way to attach labels to data. Pandas solves this by introducing an Index, allowing you to work with data exactly like a spreadsheet or an SQL database.

1. The Series (1D Data)
A Series is a one-dimensional array-like object. Think of it as a single column in an Excel sheet. It contains two main components:

Values: A sequence of data (similar to a NumPy array).

Index: An associated array of data labels.

#### Creating a Basic Series :

When you create a Series without specifying an index, pandas automatically assigns a default integer index starting from 0.

```python
import pandas as pd 
# Creating a Series from a simple list
obj  = pd.Series([4,7,-5,3])
print(obj)
```

 #### Customizing the Index:

The real power of a Series emerges when you assign meaningful labels to your data points. This allows you to select data by its label instead of its numerical position.

```python
# Creating a Series with a custom index:
sales = pd.Series([400, 700, 500, 300] , index = ['Jan' , 'Feb' , 'Mar', 'Apr'])
print(sales)

# Selecting data using the label:
print(sales['Feb'])   # Output: 700
```

#### Series as a Dictionary: 

Because a Series maps an index label to a value, you can actually think of it as a fixed-length, ordered Python dictionary. In fact, you can create a Series directly from a dictionary!

```python
Marks = {'Yashraj' : 40 , 'Annu' : 90 , 'Aman' : 70} 
Marks_series =  pd.Series(Marks)
print(Marks_series)
```

 #### 2. The DataFrame (2D Tabular Data): 
A DataFrame is a two-dimensional, rectangular table of data. Think of it as an entire Excel spreadsheet. It contains an ordered collection of columns, where each column can be a different value type (numeric, string, boolean, etc.).

Under the hood, a DataFrame is essentially a dictionary of Series objects that all share the same index.

Creating a DataFrame

The most common way to create a DataFrame from scratch is using a dictionary of equal-length lists or NumPy arrays.

```python
data = {
    'state' : ['Rajasthan' , 'Rajasthan' , 'Haryana' , 'Kerela' , 'Madhya_Pradesh'],
    'year' : [ 2000 , 2001 , 2002 , 2003 , 2004 ],
    'name' : ['Yashraj' , 'Rahul' , 'Sahil' , 'Utkarsh' , 'halwa']
}

df = pd.DataFrame(data)
print(df)
```

Key DataFrame Methods :

df.head(): Returns the first 5 rows (invaluable for large datasets).

Selecting a Column: You can extract a single column as a Series using dictionary-like notation (df['state']) or attribute notation (df.state).

```python
df.head()
```

```python
print(df['state'])
df.state
```

### 3. Index Objects

Both Series and DataFrames utilize Index objects to hold axis labels (and other metadata like axis names).

The most critical rule regarding pandas Index objects is that they are immutable (they cannot be modified by the user once created).

```python
my_index = pd.Index(['a','b','c'])
# my_index[1] = 'd'    <-- THIS WILL THROW A TYPE ERROR!
```

Why Immutability Matters: Because indexes cannot be changed, it is completely safe to share them between different DataFrames and Series without worrying about accidental modifications corrupting your data structure.

## Lecture 2 -> Essential Functionality (Manipulation & Selection).

Now that you know how to build a Series and a DataFrame, you need to know how to surgically extract the exact data you want, or get rid of the data you don't. This is where pandas truly shines.

#### 1. Dropping Entries (Cleaning House)
Often, you'll have rows or columns you simply don't need. The drop method returns a new object with the specified values removed from an axis.

```python
# Let's create a quick Series
import numpy as np 
obj = pd.Series(np.arange(5.) , index = ['a' , 'b' , 'c' , 'd' , 'e'])
print(obj)

# Drop rows 'c' and 'd'
new_obj = obj.drop(['c' , 'd'])
print(new_obj)
```

Dropping Columns (using axis=1):

When working with a DataFrame, if you want to drop a column instead of a row, you must explicitly tell pandas to look at the columns by passing axis=1 (or axis='columns').

### Examples :

```python
df = pd.DataFrame(
    np.arange(20).reshape(5,4),
    index = ['a', 'b' ,'c' , 'd' ,'e'],
    columns = ['one', 'two' ,'three' ,'four']

)
print(df)
```

```python
# Drop One Column:
new_df = df.drop('two' , axis = 1)  #axis=1 → columns (across)
print(new_df)
```

```python
# Drop Multiple Columns
new_df = df.drop(['two' ,'three'] , axis = 1)
print(new_df)
```

```python
# Using columns= Instead of axis=1
new_df = df.drop(columns = ['two' ,'three'])
print(new_df)
```

```python
#Using index= for Rows
new_df = df.drop(index = ['a', 'd'])
print(new_df)
```

 #### 2. The Golden Rule of Selection: .loc vs .iloc
If there is one thing you memorize from Chapter 5, make it this. Selecting data in pandas can be confusing because you can select by a label (like 'Ohio') or by an integer position (like row 0).

Pandas created two specific operators to make this crystal clear:

.loc (Label-based): You ask for data based on the names of the index or columns.

.iloc (Integer-based): You ask for data based on its numerical position (starting from 0), exactly like standard Python lists.

```python
data = pd.DataFrame(np.arange(9).reshape((3, 3)), 
                    index=['Ohio', 'Colorado', 'Utah'], 
                    columns=['one', 'two', 'three'])

print(data)

# 1. Grab Colorado's data using its label (.loc):
print(data.loc['Colorado'])

# 2. Grab Colorado's data using its row number (.iloc):
print(data.iloc[1])

# 3. Grab a specific slice (Rows Colorado & Utah, Columns 'two' & 'three')
print(data.loc['Colorado':'Utah' , 'two':'three'])
```

Important Note:

 When slicing with .loc (labels), the endpoint is inclusive (it includes 'Utah'). When slicing with .iloc (integers) or standard Python, the endpoint is exclusive.

###  3. Filtering (Boolean Indexing)
You will rarely just say "give me row 5". Instead, you'll ask questions like "give me all rows where the value in column 'three' is greater than 5". You do this by passing a condition directly into the brackets.

```python
filtered_data = data[data['three']>=5]
print(filtered_data)      # exmaple of single condition 
```

### Example:


```python
import pandas as pd

data = pd.DataFrame({
    'one':   [1, 2, 3, 4, 5],
    'two':   [10, 20, 30, 40, 50],
    'three': [2, 7, 4, 9, 6]
})

print(data)
```

```python
# 8. Filter Specific Columns After Filtering:
# Rows where three > 5, but show only one and three.
data.loc[data['three'] > 5, ['one', 'three']]
```

## Lecture 3: Essential Functionality (Operations).

Now that you know how to build, slice, and filter DataFrames, it is time to make them do some actual work. In this lecture, we will cover how pandas handles math between mismatched datasets, how to apply your own custom functions, and how to get your data organized.

#### 1. Arithmetic and Data Alignment:
One of the most powerful (and sometimes surprising) features of pandas is how it handles arithmetic between objects with different indexes.

If you add two Series or DataFrames together, pandas does not just blindly add row 1 to row 1. It looks at the index labels and aligns the data first. If a label exists in one dataset but not the other, pandas introduces a NaN (Not a Number), which represents missing data.

```python
s1 = pd.Series([7.3,-2.5,3.4,1.5] , index = ['a','b','c','d'])
s2 = pd.Series([-2.1, 3.6, -1.5, 4, 3.1], index=['a', 'c', 'e', 'f', 'g'])

# Adding them together:
result = s1 + s2
print(result)
```

Notice how d, f, and g became NaN. They didn't have a partner to add with, and in pandas, any math operation involving a NaN results in another NaN.

The Solution: Using Arithmetic Methods with Fill Values:

To prevent missing data from ruining your calculations, use the built-in arithmetic methods (.add(), .sub(), .div(), .mul()) and pass a fill_value.

### Examples: 

```python
# Using fill_value:
result = s1.add(s2 ,fill_value =0 )
print(result)
```

```python
# Subtraction Between Series:
s1 = pd.Series([100, 200, 300], index=['A', 'B', 'C'])
s2 = pd.Series([50, 75, 25], index=['B', 'C', 'D'])
print(s1 - s2)
result = s1.sub(s2 , fill_value = 0)
print(result)
```

```python
# DataFrame Addition:
df1 = pd.DataFrame(
    [[1, 2], [3, 4]],
    index=['row1', 'row2'],
    columns=['A', 'B']
)
print(df1)

df2 = pd.DataFrame(
    [[10, 20], [30, 40]],
    index=['row2', 'row3'],
    columns=['B', 'C']
)
print(df2)

print(df1 + df2)
```

```python
 # DataFrame with fill_value:
result = df1.add(df2 , fill_value = 0)
print(result)
```

```python
# Multiplication:
s1 = pd.Series([2, 4, 6], index=['a', 'b', 'c'])
s2 = pd.Series([10, 20, 30], index=['b', 'c', 'd'])

print(s1 * s2)
print(s1/s2)
```

#### 2. Function Application and Mapping:

NumPy functions (like np.abs() to get absolute values) work perfectly fine on pandas objects. However, you will often want to apply your own custom functions to your data.

Using .apply() on DataFrames:

The .apply() method allows you to run a function on an entire row or column at once. It is incredibly common to use Python lambda (anonymous) functions here for quick calculations.

```python
df = pd.DataFrame(
    np.random.randn(4,3),
    columns=['A', 'B', 'C'], 
    index=['Utah', 'Ohio', 'Texas', 'Oregon']

)
print(df )
## Calculate the range (max minus min) for each COLUMN:
range_func = lambda x: x.max() - x.min()
df.apply(range_func)
```

Using .map() on a Series:

If you want to transform every single individual element in a Series, you use .map(). This is highly useful for cleaning text or mapping dictionary values.

```python
# Mapping a formatting function to every element in a Series:
format_func = lambda x: f"{x:.2f}"
formatted_series = df['A'].map(format_func)
print(formatted_series)
```

#### 3. Sorting and Ranking:

Sorting datasets by an essential criterion is a daily task in data analysis. Pandas gives you two main ways to sort: by the index labels, or by the actual data values.

Sorting by Index (.sort_index())

This returns a new, sorted object based on the row or column labels.

```python
unsorted_df = pd.DataFrame(np.arange(4).reshape((2, 2)), 
                           index=['three', 'one'], 
                           columns=['d', 'a'])
print(unsorted_df)
# Sort rows alphabetically by their index labels :
print(unsorted_df.sort_index())
```

| Method                             | What it sorts                  |
| ---------------------------------- | ------------------------------ |
| `df.sort_index()`                  | Row labels (index)             |
| `df.sort_index(axis=1)`            | Column labels                  |
| `df.sort_index(ascending=False)`   | Reverse order                  |
| `df.sort_values(by='column_name')` | Actual data values in a column |


```python
print(unsorted_df.sort_index(axis=1))
print(unsorted_df.sort_index(ascending =False))
```

Sorting by Values (.sort_values()):

This is the one you will use most often. You can sort a DataFrame by the data in one or more specific columns

```python
data = pd.DataFrame({'b': [4, 7, -3, 2], 'a': [0, 1, 0, 1]})
print(data)
sorted_data = data.sort_values(by = "b")
print(sorted_data['b'])
```

Pro-tip: If you want to sort from highest to lowest, simply add ascending=False inside the parentheses: data.sort_values(by='b', ascending=False).

```python
data.sort_values(by='b', ascending=False)
```

2. Correlation and Covariance

In data analysis, you often want to know if two columns are related. For example, if advertising spend goes up, do sales go up too?

Correlation (.corr()): Measures how closely two variables move together on a scale of -1 to 1.

Covariance (.cov()): Measures the directional relationship between the variables (unscaled).

### Example:


```python
import pandas as pd

df = pd.DataFrame({
    'Advertising': [10, 20, 30, 40, 50],
    'Sales': [100, 120, 140, 160, 180]
})

print(df)
```

```python
#Correlation (corr()):
print(df['Advertising'].corr(df['Sales']))
```

```python
# Negative Correlation:
df = pd.DataFrame({
    'Exercise_Hours': [1, 2, 3, 4, 5],
    'Body_Fat': [30, 25, 20, 15, 10]
})

print(df['Exercise_Hours'].corr(df['Body_Fat']))
```

```python
#Covariance (cov())
# Covariance tells us whether variables move together.
df = pd.DataFrame({
    'Advertising': [10, 20, 30, 40, 50],
    'Sales': [100, 120, 140, 160, 180]
})
print(df['Advertising'].cov(df['Sales']))
```

| Feature           | Correlation (`corr`) | Covariance (`cov`) |
| ----------------- | -------------------- | ------------------ |
| Range             | -1 to 1              | No fixed range     |
| Shows Direction   | Yes                  | Yes                |
| Shows Strength    | Yes                  | Not directly       |
| Affected by Units | No                   | Yes                |
| Commonly Used     | Very often           | Less often         |
